```python
def search_for_datafiles():
    path = os.getcwd()

    files = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if '.json' in file:
                #files.append(os.path.join(r, file))
                files.append(file)
    
    return files
```
